```python
f = lambda s: datetime.strptime(s, '%Y%m%d')

def get_full_series(ts_code):
    today = datetime.strftime(datetime.today(), '%Y%m%d')
    start_date = df_etf[df_etf['ts_code'] == ts_code]['list_date'].values[0]
    
    full_series_list = []
    current_load = pro.fund_daily(ts_code=ts_code, start_date=start_date, end_date=today)
    full_series_list.append(current_load)
    
    while current_load['trade_date'].values[-1] != start_date:
        new_end_date = datetime.strptime(current_load['trade_date'].values[-1], '%Y%m%d') - timedelta(days=1)
        new_end_date_str = datetime.strftime(new_end_date, '%Y%m%d')
        current_load = pro.fund_daily(ts_code=ts_code, start_date=start_date, end_date=new_end_date_str)
        full_series_list.append(current_load)
        
    full_series = pd.concat(full_series_list)
    full_series['trade_date'] = full_series['trade_date'].apply(f)
    full_series.index = full_series['trade_date']
    full_series.index.name = 'Date'
    return full_series
```
